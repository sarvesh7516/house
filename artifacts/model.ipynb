```python
def convert_range_to_num(sqft_range):
    num = sqft_range.split('-')
    if len(num) == 2 :
        avg = (float(num[0]) + float(num[1])) / 2
        return avg
    try:
        return float(sqft_range)
    except:
        return None
    
        
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_squared_error
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms = MinMaxScaler()
x_train = ms.fit_transform(x_train)
x_test = ms.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
l_reg = LinearRegression()
l_reg.fit(x_train,y_train)

y_pred = l_reg.predict(x_test)

l_acc = r2_score(y_test,y_pred)
l_mse = mean_squared_error(y_test,y_pred)
print(f'Accuracy={l_acc}')
print(f'MSE={l_mse}')
```

```python
k_reg = KNeighborsRegressor()
k_reg.fit(x_train,y_train)

y_pred_k = k_reg.predict(x_test)

k_acc = r2_score(y_test,y_pred_k)
k_mse = mean_squared_error(y_test,y_pred_k)
print(f'Accuracy={k_acc}')
print(f'MSE={k_mse}')
```

```python
d_reg = DecisionTreeRegressor()
d_reg.fit(x_train,y_train)

y_pred_d = d_reg.predict(x_test)

d_acc = r2_score(y_test,y_pred_d)
d_mse = mean_squared_error(y_test,y_pred_d)
print(f'Accuracy={d_acc}')
print(f'MSE={d_mse}')
```

```python
r_reg = RandomForestRegressor()
r_reg.fit(x_train,y_train)

y_pred_r = r_reg.predict(x_test)

r_acc = r2_score(y_test,y_pred_r)
r_mse = mean_squared_error(y_test,y_pred_r)
print(f'Accuracy={r_acc}')
print(f'MSE={r_mse}')
```

```python
%%time
x_reg = XGBRegressor()
x_reg.fit(x_train,y_train)

y_pred_x = x_reg.predict(x_test)

x_acc = r2_score(y_test,y_pred_x)
x_mse = mean_squared_error(y_test,y_pred_x)
print(f'Accuracy={x_acc}')
print(f'MSE={x_mse}')
```

```python
models = pd.DataFrame({
                        'Model':['Linear','KNN','Decision Tree','Random Forest','XGBOOST'],
                        'Accuracy':[l_acc,k_acc,d_acc,r_acc,x_acc] })

models
```
